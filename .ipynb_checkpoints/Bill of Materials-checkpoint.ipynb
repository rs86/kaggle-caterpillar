```python
import pandas as pd
```

```python
df = pd.read_csv('data/bill_of_materials.csv')
```

```python
assemblies = []
for i in range(50):#range(len(df)):
    r = dict(df.loc[i])
    ta = r['tube_assembly_id']
    for i in [1, 2, 3, 4, 5, 6, 7, 8]:
        key_component_id = 'component_id_' + str(i)
        if type(r[key_component_id]) == str:
            component_id = r[key_component_id]
            quantity = r['quantity_' + str(i)]
            assemblies.append((ta, component_id, quantity))            
```

```python
df = pd.DataFrame(assemblies, columns=['tube_assembly_id', 'component_id', 'quantity'])
```

```python
pd.pivot_table(df, index='tube_assembly_id', columns='component_id', fill_value=0)
```
